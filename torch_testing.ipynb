```python
d = dict([(name, cls) for name, cls in torch.nn.__dict__.items() if isinstance(cls, type)])
od = dict([(name, cls) for name, cls in optim.__dict__.items() if isinstance(cls, type)])
```

```python
class NetworkGraph():
  def __init__(self, nodes):
    """
    nodes -> nodes
    edges -> connectivity
    edge_attrs -> activation fns, aggregation rules
    """
    self.nodes = nodes
   

class Node():
  def __init__(self, v):  # args, kwargs must be empty iterable, map, respectively
    self.name, self.fn, self.args, self.kwargs = v

  def parse(self, d):
    return d[self.fn](*self.args, **self.kwargs)


class FlexNet(torch.nn.Module):
  def __init__(self, graph, fns=d):  # G = (V, E)
    super().__init__()
    self.graph = graph
    self.d = torch.nn.ModuleDict({node.name: node.parse(d) for node in graph.nodes})

  def forward(self, x):
    for layer in self.d.values():
      x = layer(x)
    return x


#  edge list -> always start with (input, layer)
#  corresponding attr applies function to out node's parents
```

```python
from torchvision import datasets, transforms

transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

trainset = datasets.MNIST('PATH_TO_STORE_TRAINSET', download=True, train=True, transform=transform)
valset = datasets.MNIST('PATH_TO_STORE_TESTSET', download=True, train=False, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
valloader = torch.utils.data.DataLoader(valset, batch_size=64, shuffle=True)
```

```python
import ast
import inspect 

layers = []
while True:
  name = input('enter layer name, or type "esc" to escape ')
  if name == 'esc':
    break
  print(d)
  layer = input('enter pytorch layer ')
  print(inspect.getfullargspec(d[layer]))
  args = input('enter comma-separated integer args ')
  args = map(int, args.split(','))
  #kwargs = input('enter key: value pairs of kwargs ')
  #kwargs = ast.literal_eval(f'\{{kwargs}}')
  kwargs = {}
  layers.append(Node([name, layer, args, kwargs]))

net = NetworkGraph(layers)
fnet = FlexNet(net)

print(d)
loss = d[input("enter loss fn ")]()

print(od)
opt = od[input("enter optimization algorithm ")]
lr = int(input("enter lr "))
optimizer = opt(net.parameters(), lr=lr)


epochs = input("enter epochs ")
```

```python
for e in range(epochs):
    running_loss = 0
    for images, labels in trainloader:
        images = images.view(images.shape[0], -1)
    
        # Training pass
        optimizer.zero_grad()
        
        output = net(images)
        loss = loss(output, labels)
        
        #This is where the model learns by backpropagating
        loss.backward()
        
        #And optimizes its weights here
        optimizer.step()
        
        running_loss += loss.item()
    else:
        print("Epoch {} - Training loss: {}".format(e, running_loss/len(trainloader)))
```
